```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Lat2.xlsx')
df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#-- Pilih fiture numerik --
fiture=['pendapatan','pengeluaran','usia']
X=df[fiture]
X.head()
```

```python
#-- Menggunakan Normalisasi Data Menggunakan STD--
scaler=StandardScaler()
X_std=scaler.fit_transform(X)
X_std
```

```python
kmeans=KMeans(n_clusters=3, random_state=42,n_init='auto')
df['Cluster']=kmeans.fit_predict(X)
df.head(40)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 2D Scatter Plot (Pendapatan vs Pengeluaran)
plt.figure(figsize=(6,4))
plt.scatter(df['pendapatan'], df['pengeluaran'], c=df['Cluster'], cmap='viridis', s=100)
plt.xlabel('Pendapatan')
plt.ylabel('Pengeluaran')
plt.title('Cluster Berdasarkan Pendapatan dan Pengeluaran')
plt.grid(True)
plt.colorbar(label='Cluster')
plt.show()

# 3D Scatter Plot (Pendapatan vs Pengeluaran vs Usia)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df['pendapatan'], df['pengeluaran'], df['usia'],
                     c=df['Cluster'], cmap='viridis', s=100)
ax.set_xlabel('Pendapatan')
ax.set_ylabel('Pengeluaran')
ax.set_zlabel('Usia')
ax.set_title('3D Cluster Visualisasi: Pendapatan, Pengeluaran, Usia')

legend1 = ax.legend(*scatter.legend_elements(), title="Cluster")
ax.add_artist(legend1)
plt.show()
```
