```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('orders_data.xlsx')
df.head()
```

```python
#-- Melihat berapa banyak data yang null (kosong)--
df.isnull().sum()
```

```python
#-- Melihat banyak data dan banyak column --
df.shape
```

```python
df.info()
```

```python
df['item_total'].head()
```

```python
#-- buat list value yang kosong --
value=[]
#-- ulang variabel u dengan data item_total
for u in df['item_total']:
  #-- Jika ada data u yang null maka isi value= 0.0 kalau tidak
  #-- data u yang ada mata uang dihapus dan yang ada koma di hilangkan --
  if pd.isna(u):
    value.append(0.0)
  else:
    c=float(u.replace('₹','').replace(',',''))
    #-- tambahkan data di variabel c kedalam list value --
    value.append(c)
df['item_total']=value

```

```python
value=[]
for u in df['shipping_fee']:
  if pd.isna(u):
    value.append(0.0)
  else:
    c=float(u.replace('₹','').replace(',',''))
    value.append(c)
df['shipping_fee']=value
```
