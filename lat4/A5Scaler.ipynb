```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=pd.read_csv('fruit_data_with_colours.csv')
dt.head()
```

```python
#-- Melihat data yang. null --
dt.isnull().sum()
```

```python
X=dt[['height','width','mass','color_score']]
y=dt['fruit_label']
```

```python
X
```

```python
from sklearn.preprocessing import MinMaxScaler
st=MinMaxScaler()
dts=st.fit_transform(X)
```

```python
dts
```

```python
dt.dtypes
```

```python
from sklearn.model_selection import train_test_split
x_train,x_tes, y_train, y_tes=train_test_split(dts,y, test_size=0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_tes, y_tes)
```
