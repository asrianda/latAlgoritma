```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[2,3,4,2,7,5]
y=[3,4,1,3,1,5]
plt.figure(figsize=(5,4))
#plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.cluster import KMeans
data=list(zip(x,y))
print(data)

kluster=[]
for i in range(1,6):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(data)
    kluster.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),kluster,marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('kluster')
plt.show()
```
